```python
# prompt: from the data in the file give me  linear regression graph code between the TS135 and TS144  in python and also represent the title in that graph

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Extract the relevant columns
x = df['TS135'].values.reshape(-1, 1)
y = df['TS144'].values

# Create a linear regression model
model = LinearRegression()
model.fit(x, y)

# Predict the values
y_pred = model.predict(x)

# Plot the data and the regression line
plt.scatter(x, y, color='blue')
plt.plot(x, y_pred, color='red')

# Add labels and title
plt.xlabel('TS135')
plt.ylabel('TS144')
plt.title('Linear Regression between TS135 and TS144')

# Show the plot
plt.show()
```
